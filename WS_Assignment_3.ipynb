2. In the above question, now scrape the following details of each product listed in first 3 pages of your search 
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 


```python
from selenium.common.exceptions import NoSuchElementException

product_url = []
url = driver.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]/a')

    
for i in url:
    product_url.append(i.get_attribute('href'))


price= []
rep = [] 
brand = []
delivery = []
availablity = []
product_name = []

#by using try and except we can scrap all the details by clicking on product url
for i in product_url:
    driver.get(i)
    time.sleep(2)
    try:
        prices = driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]/span/span[2]')
        price.append(prices.text.replace('\n00',""))
    except NoSuchElementException:
        price.append('-')

    try:
        replacement = driver.find_element(By.XPATH,"/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span")
        rep.append(replacement.text)
    except NoSuchElementException:
        rep.append('-')    

    try:
        brands = driver.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]/td[2]/span')
        brand.append(brands.text)
    except NoSuchElementException:
        brand.append('-')
        
    try:
        deli = driver.find_element(By.XPATH,"/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span")
        delivery.append('Yes')
    except NoSuchElementException:
        delivery.append('Not Present')
        
    try:
        avail = driver.find_element(By.XPATH,'//div[@id="availability"]/span')
        availablity.append(avail.text)
    except NoSuchElementException:
        availablity.append('-')
        
    try:
        name=driver.find_element(By.XPATH,'//h1[@id="title"]/span')
        prod_name=name.text
        product_name.append(prod_name)
        if(prod_name == 0):
            try:
                name1=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[6]/div[4]/div[4]/div[1]/div/h1/span')
                product_name.append(name1.text)
            except NoSuchElementException:
                product_name.append("-")
    except:
        product_name.append("-")
        
driver.close()

#dataframe
product = pd.DataFrame({'Product Name':product_name,'Brand':brand,'Price':price,'Availability':availablity,'Delivery':delivery,'Replacement/Exchange':rep,'URL':product_url})
product
```


4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
details is missing then replace it by “- “. Save your results in a dataframe and CSV. 


```python
driver=webdriver.Chrome()  
driver.get('https://www.flipkart.com/')

#closing the login dialog
close = driver.find_element(By.CLASS_NAME,"_30XB9F")
close.click()

brand=input("Enter smartphone name : ")

serch_bar = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/header/div[1]/div[2]/form/div/div/input')
serch_bar.send_keys(brand)

# Clicking on search icon
search_button=driver.find_element(By.CLASS_NAME,'_2iLD__')
search_button.click()


time.sleep(5)

brand_titles=[]
product_price=[]
display_size=[]
ram=[]
rom=[]
product_color=[]
battery_size=[]
product_camera=[]
product_url=[]

#Brands
Brand_tags=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in Brand_tags:
    brand_titles.append(i.text[0:11])

#Prices
Brand_price=driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
for i in Brand_price:
    product_price.append(i.text)

#Display size
display_tags=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]/li[2]')
for i in display_tags:
    display_size.append(i.text)    

#RAm    
ram_p=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
for i in ram_p:
    ram.append(i.text[0:9])

#ROM   
rom_p=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
for i in rom_p:
    rom.append(i.text[11:17])    
    
#COLOR
color_p=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in color_p:
    product_color.append(i.text[11:25])
    

#Battery size
battery_p=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
for i in battery_p:
    battery_size.append(i.text.splitlines()[3])

#Camera    
camera_p=driver.find_elements(By.XPATH,'//ul[@class="_1xgFaf"]')
for i in camera_p:
    product_camera.append(i.text.splitlines()[2])

rear_cameras = []
front_cameras = []

for specs in product_camera:
    if '|' in specs:
        rear, front = specs.split('|')
        rear_cameras.append(rear.strip())  
        front_cameras.append(front.strip())
    else:
        rear_cameras.append(specs)
        front_cameras.append('-')    
    
url_tags=driver.find_elements(By.CLASS_NAME,'_1fQZEK')
for i in url_tags:
    product_url.append(i.get_attribute('href'))
    
# Dataframe   
df=pd.DataFrame({'Brand Names':brand_titles,'Product Price':product_price,'Display size':display_size,'Ram Size ( GB )':ram,'Rom size':rom,'SmartPhone Color':product_color,'SmartPhone Battery Size':battery_size,'Primary Camera':rear_cameras,'Secondary Camera':front_cameras,'Product Link':product_url})

#save to csv file
df.to_csv('//FlipRobo/Smartphone.csv')
```


5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 


```python
driver=webdriver.Chrome()  
driver.get("https://www.google.com/maps/")

# Desired location from the user
user_input=input("Enter city name for which coordinates is required : ")


search=driver.find_element(By.XPATH,'//input[@autofocus="autofocus"]')
search.click()
search.send_keys(user_input)

search_button=driver.find_element(By.ID,'searchbox-searchbutton')
search_button.click()
time.sleep(3)

lat=[]
long=[]
try:
    url_string = driver.current_url
    print("URL Extracted :",url_string)
    lat_long = re.findall(r'@(.*)data',url_string)
    if len(lat_long):
        lat_long_list = lat_long[0].split(",")
        if len(lat_long_list)>=2:
            lat.append(lat_long_list[0])
            long.append(lat_long_list[1])
except Exception as e:
    print("error:",str(e))


# displaying the coordinates 
location_data=pd.DataFrame({'Enter Location':user_input,'latitude':lat,'longitude':long})
location_data
```


8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted 
from any YouTube Video. 


```python
#https://www.youtube.com/watch?v=e60rJHBw_CM

url=input("Enter the URL: ")
driver=webdriver.Chrome()  
driver.get(url)
time.sleep(5)

comments = []
upvote = []
posted = []

for _ in range(30):
    driver.execute_script("window.scrollBy(0,100)")
    time.sleep(3)
    
    com_xpath = driver.find_elements(By.ID,"content-text")
    for i in com_xpath:
        comments.append(i.text)
           
        
    post = driver.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
    for i in post:
        posted.append(i.text)
        
        
    upvotes = driver.find_elements(By.ID,"vote-count-middle")
    for i in upvotes:
        upvote.append(i.text)

     

df=pd.DataFrame({'Comments':comments,'Comments Upvote':upvote,'Comment is POSTED':posted})
df
```


9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall 
reviews, privates from price, dorms from price, facilities and property description. 
 

```python
driver=webdriver.Chrome()
try:

    driver.get("https://www.hostelworld.com/")

    loc = driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input")
    loc.send_keys("london")

    driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div[1]/div/div[2]/div/ul/li[2]/button').click()


except Exception as e:

    driver.get("https://www.hostelworld.com/s?q=London, England&country=England&city=London&type=city&id=3&from=2022-07-28&to=2022-07-31&guests=2&HostelNumber")


time.sleep(3)



hostel_name = []
location = []
ratings = []
total_reviews = []
overall_reviews = []
private_price = []
dorm_price = []
property_description=[]

hostel = driver.find_elements(By.XPATH, '//div[@class="property-name"]/span')
try:
    for i in hostel:
        hostel_name.append(i.text)
except:
    hostel_name.append("-")


distance = driver.find_elements(By.XPATH, '//span[@class="distance-description"]')
try:
    for i in distance:
        location.append(i.text)
except:
    location.append("-")


information = driver.find_elements(By.XPATH, '//span[@class="number"]')
try:
    for i in information[2:]:
        ratings.append(i.text.split("\n")[0])
except:
    ratings.append("-")

reviews = driver.find_elements(By.XPATH, "//div[@class='review']/span")
try:
    for i in reviews:
        total_reviews.append(i.text)
except:
    total_reviews.append("-")


review = driver.find_elements(By.XPATH, '//span[@class="keyword"]')
try:
    for i in review[2:]:
        overall_reviews.append(i.text)
except:
    overall_reviews.append("-")


private = driver.find_elements(By.XPATH, '//div[@class="property-accommodation-prices"]/div[1]')
try:
    for i in private:
        private_price.append(i.text.replace("\n", ""))
except:
    private_price.append("-")


dorm = driver.find_elements(By.XPATH, '//div[@class="property-accommodation-prices"]/div[1]')
for i in dorm:
    dorm_price.append(i.text.replace("\n", ""))


property_description = []
prop_url = []
url = driver.find_elements(By.XPATH, '//div[@class="property-card"]/a')
for i in url[2:]:
    prop_url.append(i.get_attribute("href"))

for i in prop_url:
    driver.get(i)

    time.sleep(2)
    try:
        description = driver.find_element(
            By.XPATH,
            '//div[@class="description-container"]/div/div[2]',
        )
        property_description.append(description.text)
    except NoSuchElementException:
        property_description.append("-")



available_hostel = pd.DataFrame(
    {
        "Hostel Name": hostel_name,
        "Distance from City Center": location,
        "Ratings": ratings,
        "Total Review": total_reviews,
        "Overall Review": overall_reviews,
        "Private Price": private_price,
        "Dorm Price": dorm_price,
        "Description": property_description,
        "URL": prop_url,
    }
)

available_hostel
```
