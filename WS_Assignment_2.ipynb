*Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.\
This task will be done in following steps:\
1.First get the webpage https://www.shine.com/\
2.Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.\
3.Then click the searchbutton.\
4.Then scrape the data for the first 10 jobs results you get.\
5.Finally create a dataframe of the scraped data.*

```python
# Importing Libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


#setting up the webdriver
driver=webdriver.Chrome()
driver.get('https://www.shine.com/new/job-search')

# Entering "Data Analyst" as the job title
designation=driver.find_element(By.CLASS_NAME,"form-control  ")
designation.send_keys("Data Analyst")

#Entering "Bangalore" as job location
location = driver.find_element(By.ID,'id_loc')
location.send_keys("Bangalore")

#Clicking on "Search Jobs" button
search_button = driver.find_element(By.XPATH, "/html/body/div/div[1]/div/div/div[1]/div[1]/div/div[2]/div/div/form/div/div[2]/div/button")
search_button.click()


#Scraping the data for Data Analyst job roles in Bangalore location

job_title=[]
job_loc=[]
exp_required=[]
company_name=[]

# Scraping the job titles
job_tags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')
for i in job_tags[:10]:
    job_title.append(i.text)
   

# Scraping the job Location
loc_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in loc_tags[:10]:
    job_loc.append(i.text)

# Scraping the Company's Name
company_tags=driver.find_elements(By.CLASS_NAME,"jobCard_jobCard_cName__mYnow")
for i in company_tags[:10]:
    company_name.append(i.text)
    
# Scraping the Experience required
exp_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_lists__fdnsc"]/div[1]')
for i in exp_tags[:10]:
    exp_required.append(i.text)

# Creating the Dataframe
df=pd.DataFrame({"Job_title":job_title,"Location":job_loc,"Company":company_name,"Experience_Needed":exp_required})
df
```

*Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.\
The above task will be done in following steps:\
1.First get the webpage https://www.azquotes.com/\
2.Click on Top Quotes\
3.Then scrap a) Quote b) Author c) Type Of Quotes*

```python
#Importing libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import time
```

```python
# setting up the webdriver
driver=webdriver.Chrome()
driver.get("https://www.azquotes.com/")


# clicking on "TOP QUOTES"
top_quotes=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top_quotes.click()

# Scraping data

quote=[]
author=[]
type_of_quote=[]

# Scraping the quotes
for page in range(0,10):
    quote_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for q in quote_tag:
        quote.append(q.text)
    if page != 9:    
        next_button=driver.find_element(By.CLASS_NAME,'next')
        next_button.click()
        time.sleep(2)      
    
```

```python
#Opening the 1st page
driver.get("https://www.azquotes.com/top_quotes.html")

# Scraping the authors
for page in range(0,10):
    author_tag=driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    for a in author_tag:
        author.append(a.text)
    if page != 9:    
        next_button=driver.find_element(By.CLASS_NAME,'next')
        next_button.click()
        time.sleep(2)            

```

```python
#Opening the 1st page
driver.get("https://www.azquotes.com/top_quotes.html")

# Scraping the type of quote
for page in range(0,10):
    type_tag=driver.find_elements(By.CLASS_NAME,'tags')
    for t in type_tag:
        type_of_quote.append(t.text)
    if page != 9:    
        next_button=driver.find_element(By.CLASS_NAME,'next')
        next_button.click()
        time.sleep(2)            
```

```python
# creating the dataframe
df=pd.DataFrame({"Quote":quote,"Author":author,"Type Of Quote":type_of_quote})
df
```

```python
# Importing Libraries
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import time



#setting up the webdriver
driver=webdriver.Chrome()
driver.get('https://www.motor1.com/')

time.sleep(1)

# Entering "50 most expensive cars" keyword
keyword=driver.find_element(By.ID,'search_input')
keyword.send_keys("50 most expensive cars")



# Clicking on search icon
search_button=driver.find_element(By.XPATH,'//*[@id="header_search_form"]/button[1]')
search_button.click()

time.sleep(1)

# opening the list for 50 most expensive cars
list=driver.find_element(By.XPATH,'/html/body/div[10]/div[9]/div/div[1]/div/div/div[2]/div/div[1]/a').click()

time.sleep(0.5)
```

```python
# scarping data

cars=[]
price=[]

#scarping car names
car_tags=driver.find_elements(By.CLASS_NAME,'subheader')
for c in car_tags[:50]:
    cars.append(c.text)

# scraping price
price_tags=driver.find_elements(By.TAG_NAME,'strong')
for p in price_tags:
    price.append(p.text)


#creating dataframe
df=pd.DataFrame({"Car":cars,"Price":price})
df
```
